```python
import os
import csv


PyPollcsv = os.path.join("..","Resources","election_data.csv")


count_votes = 0
candidate_list = []
unique_candidate = []
vote_count = []
vote_percent = []


with open(PyPollcsv, newline="") as csvfile:
    
    csvreader = csv.reader(csvfile, delimiter=",")
    csv_header = next(csvreader)
   
    
    for row in csvreader:
        
        count_votes = count_votes + 1
     
        candidate_list.append(row[2])
       
    for x in set(candidate_list):
        unique_candidate.append(x)
      
        y = candidate_list.count(x)
        vote_count.append(y)
       
        z = round((y/count_votes)*100,3)
        vote_percent.append(z)
        
    winning_vote_count = max(vote_count)
    winner = unique_candidate[vote_count.index(winning_vote_count)]
    

print("-------------------------")
print("Election Results")   
print("-------------------------")
print("Total Votes :" + str(count_votes))    
print("-------------------------")
for i in range(len(unique_candidate)):
            print(unique_candidate[i] + ": " + str(vote_percent[i]) +"% (" + str(vote_count[i])+ ")")
print("-------------------------")
print("The winner is: " + winner)
print("-------------------------")



election_results1 = os.path.join("..","analysis", "election_results1.txt")

with open("election_results1.txt", "w") as outfile:

    outfile.write("Election Results1\n")
    outfile.write("---------------------------------------\n")
    outfile.write("Total Vote: " + str(count_votes) + "\n")
    outfile.write("---------------------------------------\n")
    for i in range(len(set(unique_candidate))):
        outfile.write(unique_candidate[i] + ": " + str(vote_percent[i]) +"% (" + str(vote_count[i]) + ")\n")
    outfile.write("---------------------------------------\n")
    outfile.write("The winner is: " + winner + "\n")
    outfile.write("---------------------------------------\n")
```
